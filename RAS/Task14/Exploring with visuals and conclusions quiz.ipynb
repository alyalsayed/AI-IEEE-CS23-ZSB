# Drawing Conclusions

### Q1: How much have vehicle classes improved in fuel economy in 2018 compared to 2008?  

```python
# use plots if you can


mpg_08 = df_08.groupby('veh_class')['cmb_mpg'].mean()
mpg_18 = df_18.groupby('veh_class')['cmb_mpg'].mean()

pct_change = (mpg_18 - mpg_08) / mpg_08 * 100

# Create a horizontal bar chart 
plt.figure(figsize=(8, 6))
pct_change.plot(kind='barh', color='green')

# Add labels and title
plt.xlabel('Percentage Change')
plt.ylabel('Vehicle Class')
plt.title('Percentage Change in Average Combined MPG by Vehicle Class (2008-2018)')


plt.show()
```

### Q2: What are the characteristics of SmartWay vehicles? Have they changed over time?
Compare the following features --> drive, cyl, fuel in both datasets. Write your comments and findings in each cell  
Use plots

```python
# compare drive

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
#we can query method instead of boolean indexing
df_08.query('smartway == "yes"')['drive'].value_counts().plot(kind='bar', alpha=1, color='green', label='SmartWay')
df_08.query('smartway == "no"')['drive'].value_counts().plot(kind='bar', alpha=0.3, color='red', label='Non-SmartWay')
plt.xlabel('Drive Type')
plt.ylabel('Count')
plt.title('Distribution of Drive Types in 2008')
plt.legend()

plt.subplot(1, 2, 2)
df_18.query('smartway == "Yes"')['drive'].value_counts().plot(kind='bar', alpha=1, color='green', label='SmartWay')
df_18.query('smartway == "No"')['drive'].value_counts().plot(kind='bar', alpha=0.3, color='red', label='Non-SmartWay')
plt.xlabel('Drive Type')
plt.ylabel('Count')
plt.title('Distribution of Drive Types in 2018')
plt.legend()

plt.show() # it seems there is a decrease from 2008 to 2018 for both drive types for smartway vechicles
```

```python
# compare the number of cylinders
# Can you use a pie chart with the percentages shown?

# Create boolean masks
sw_08 = df_08['smartway'] == 'yes'
sw_18 = df_18['smartway'] == 'Yes'


# Calculate the distribution of number of cylinders
sw_cyl_08 = df_08[sw_08]['cyl'].value_counts()
sw_cyl_18 = df_18[sw_18]['cyl'].value_counts()


# Create pie charts for the distribution of number of cylinders in both datasets
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.pie(sw_cyl_08, labels=sw_cyl_08.index.map(str), autopct='%1.1f%%')
plt.title('Distribution of Number of Cylinders for SmartWay Vehicles in 2008')
#plt.legend(title='Number of Cylinders')

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.pie(sw_cyl_18, labels=sw_cyl_18.index.map(str), autopct='%1.1f%%')
plt.title('Distribution of Number of Cylinders in 2018')
#plt.legend(title='Number of Cylinders')

plt.show()
```

```python
# compare the fuel used

# Create boolean masks for SmartWay vehicles in both datasets
sw_08 = df_08['smartway'] == 'yes'
sw_18 = df_18['smartway'] == 'Yes'

# Calculate the average fuel consumption in 2008 and 2018
sw_fuel_08 = df_08[sw_08].groupby('fuel')['cmb_mpg'].mean()
sw_fuel_18 = df_18[sw_18].groupby('fuel')['cmb_mpg'].mean()

# Create bar charts for both datasets
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sw_fuel_08.plot(kind='bar', alpha=0.4, color='cyan')
plt.xlabel('Fuel Type')
plt.ylabel('Average Combined MPG')
plt.title('Average Fuel Consumption for SmartWay Vehicles in 2008')

plt.subplot(1, 2, 2)
sw_fuel_18.plot(kind='bar', alpha=0.4, color='green')
plt.xlabel('Fuel Type')
plt.ylabel('Average Combined MPG')
plt.title('Average Fuel Consumption for SmartWay Vehicles in 2018')

plt.show() # for smartway vechicles it seems that not all types of fuel are appeared and electricity is the most used in 2018
```

### Q6: For all of the models that were produced in 2008 that are still being produced in 2018, how much has the mpg improved and which vehicle improved the most?
This is a question regarding models that were updated since 2008 and still being produced in 2018. In order to do this, we need a way to compare models that exist in both datasets. You need to learn how to [merge](https://pandas.pydata.org/pandas-docs/stable/user_guide/merging.html) dataframes

## Merging Datasets
Rename 2008 columns to distinguish from 2018 columns after the merge. 
To do this, take the first 10 characters of the column label and and concatenate it with _2008. (Only take the first 10 characters to prevent really long column names.)

```python
# rename 2008 columns only
df_08 = df_08.rename(columns=lambda x: x[:10]+'_2008')
df_08.columns
```

```python
# merge datasets
merged_df = df_08.merge(df_18,left_on='model_2008', right_on='model',how='inner')
```
